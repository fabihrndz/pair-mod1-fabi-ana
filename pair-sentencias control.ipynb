# List Comprehension

```python
# 1. Utiliza list comprehension para generar una lista que contenga todos los números pares en el rango del 1 al 20.

numero = [n for n in range(0,21,2)]
print(numero)
```

```python
# 2 Utiliza list comprehension para crear una lista que contenga todas las vocales presentes en una cadena de texto (sin repetición).

cadena = "hola python"

lista_cadena = []
vocales = "aeiou"

lista = list(set(n for n in cadena if n in vocales))

print(lista)
```

```python
# otra opción para resolver ejercicio 2

cadena = "hola python"

lista_cadena = []
vocales = "aeiou"

lista = list({n for n in cadena if n in vocales})

print(lista)
```

```python
# 3 Utiliza list comprehension para filtrar los números positivos de una lista y calcular el cuadrado de cada uno.

lista_num = [5, 6, 7, -1, 20, -45]


lista_positivos = [x ** 2 for x in lista_num if x > 0]
lista_negativos = [x for x in lista_num if x < 0]

print(lista_positivos)
print(lista_negativos)
```

# Try... except

```python
# 1 Solicita al usuario una lista de nombres y un nombre para buscar en la lista. Intenta encontrar el nombre en la lista y utiliza las sentencias try y except para capturar la excepción en caso de que el nombre no sea encontrado.

lista_nombres = input("Introduce una lista de nombres separados por comas: ").lower()
nombre = input("Introduce un nombre a buscar: ").lower()

lista_nombres = lista_nombres.split(",")

lista_nombres = [n.strip() for n in lista_nombres]

try:
    lista_nombres.index(nombre)
    print("El nombre está en la lista")
except:
    print("El nombre no está en la lista")
```
